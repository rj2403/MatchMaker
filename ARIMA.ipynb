```python
def get_predictions(dataf,n):
    for i in range(n):
        print(dataf.shape)
        p = d = q = range(0, 2)
        pdq = list(itertools.product(p, d, q))
        seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
        aic = 10000
        p = None
        ps = None
        rs = []
        for param in pdq:
            for param_seasonal in seasonal_pdq:
                mod = sm.tsa.statespace.SARIMAX(dataf,order=param,seasonal_order=param_seasonal,
                                                    enforce_stationarity=True,
                                                    enforce_invertibility=True)
                results = mod.fit()
                if(results.aic<aic):
                    aic = results.aic
                    p = param
                    ps = param_seasonal
                rs.append('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        mod = sm.tsa.statespace.SARIMAX(dataf,
                                order=p,
                                seasonal_order=ps,
                                enforce_stationarity=False,
                                enforce_invertibility=False)
        results = mod.fit()
        date = dataf.index.tolist()[-1]
        
        date = date + relativedelta(months=1)
        date = date.date() 
        pred = results.get_prediction(start=pd.to_datetime(date), dynamic=False)
        dataf["Date"] = dataf.index
        dataf = dataf.reset_index(drop =True)
        pred_ci = pred.conf_int()
        if(pred_ci["lower y"].tolist()[0]<0):
            f = (pred_ci["lower y"].tolist()[0] + pred_ci["upper y"].tolist()[0])//2
        else:
            f = int(pred_ci["lower y"].tolist()[0])
        dataf.loc[len(dataf.index)] = [f,date]
        dataf["Date"] = pd.to_datetime(dataf["Date"])
        dataf = dataf.set_index('Date')
    return dataf
```

```python
rs = get_predictions(train,6)
```
